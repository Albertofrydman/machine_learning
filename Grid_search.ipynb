```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
diabetes=load_diabetes(as_frame=True,scaled=False)
X=diabetes.data
y=diabetes.target
```

```python
diabetes=load_diabetes()
param_grid = {
"n_neighbors" : [5,10],
"weights" : ["uniform", "distance"],
"algorithm" : ["auto", "ball_tree", "kd_tree"]
}

KNN = KNeighborsRegressor()

clf = GridSearchCV(KNN, param_grid,refit="neg_mean_squared_error",cv=3,scoring = ["neg_mean_squared_error","neg_root_mean_squared_error"])
clf.fit(diabetes.data, diabetes.target)
print(sorted(clf.cv_results_.keys()))

#Imprime los mejores parametros
"""print(clf.best_estimator_)
print(clf.best_score_)
print(clf.best_params_)
#print(clf.scorer_)"""

# Predecir con el mejor modelo
"""mejor_modelo = clf.best_estimator_
predicciones = mejor_modelo.predict(X)
predicciones"""

#GridSearchCV(KNN, param_grid,refit="neg_mean_squared_error",cv=3,scoring = ["neg_mean_squared_error","neg_root_mean_squared_error"])
#GridSearchCV(KNN, param_grid,refit="neg_root_mean_squared_error",cv=3,scoring = ["neg_mean_squared_error","neg_root_mean_squared_error"])

#GridSearchCV(KNN, param_grid,refit=True,cv=3)
```
