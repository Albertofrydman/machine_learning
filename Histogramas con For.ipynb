```python
import pandas as pd
import matplotlib.pyplot as plt

# Supongamos que tienes un DataFrame con tres variables categóricas
data = {
    'Categoria1': ['make'],
    'Categoria2': ['fuel_type'],
    'Categoria3': ['aspiration']
}

df = pd.DataFrame(data)

# Lista de las columnas categóricas para las que deseas crear histogramas
categorias = ['Categoria1', 'Categoria2', 'Categoria3']

# Configuración de la figura y los ejes para los histogramas
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Crear histogramas en un bucle
for i, categoria in enumerate(categorias):
    print(f"Índice: {i}, Valor: {categoria}")
    df[categoria].value_counts().plot(kind='bar', ax=axs[i])
    axs[i].set_title(f'Histograma de {categoria}')
    axs[i].set_xlabel('Categorías')
    axs[i].set_ylabel('Frecuencia')

# Ajustar el espacio entre subgráficas
plt.tight_layout()

# Mostrar los histogramas
plt.show()
```
