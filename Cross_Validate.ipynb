```python
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_validate
from sklearn.metrics import make_scorer
from sklearn.metrics import confusion_matrix
from sklearn.svm import LinearSVC
diabetes = datasets.load_diabetes()
X = diabetes.data[:150]
y = diabetes.target[:150]
linear_regressor = LinearRegression()
```

```python
cv_results = cross_validate(linear_regressor, X, y, cv=3,scoring = ["neg_mean_squared_error","r2"], return_estimator=True,return_train_score=True)
print(sorted(cv_results.keys()))
print(cv_results['train_neg_mean_squared_error'].mean())
print(cv_results['test_neg_mean_squared_error'])
```
